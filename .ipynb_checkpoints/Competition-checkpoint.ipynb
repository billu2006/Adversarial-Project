```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
import torchvision
from torchvision import transforms
from torch.autograd import Variable
import argparse
import time
import copy

# Change this id to your student id
id_ = 201808020

# setup training parameters
parser = argparse.ArgumentParser(description='PyTorch MNIST Training')
parser.add_argument('--batch-size', type=int, default=128, metavar='N',
                    help='input batch size for training (default: 128)')
parser.add_argument('--test-batch-size', type=int, default=128, metavar='N',
                    help='input batch size for testing (default: 128)')
parser.add_argument('--epochs', type=int, default=10, metavar='N',
                    help='number of epochs to train')
parser.add_argument('--lr', type=float, default=0.01, metavar='LR',
                    help='learning rate')
parser.add_argument('--no-cuda', action='store_true', default=False,
                    help='disables CUDA training')
parser.add_argument('--seed', type=int, default=1, metavar='S',
                    help='random seed (default: 1)')


args = parser.parse_args(args=[]) 

# judge cuda is available or not
use_cuda = not args.no_cuda and torch.cuda.is_available()
#device = torch.device("cuda" if use_cuda else "cpu")
device = torch.device("cpu")

torch.manual_seed(args.seed)
kwargs = {'num_workers': 1, 'pin_memory': True} if use_cuda else {}
```

```python
train_set = torchvision.datasets.FashionMNIST(root='data', train=True, download=True, transform=transforms.Compose([transforms.ToTensor()]))
train_loader = DataLoader(train_set, batch_size=args.batch_size, shuffle=True)

test_set = torchvision.datasets.FashionMNIST(root='data', train=False, download=True, transform=transforms.Compose([transforms.ToTensor()]))
test_loader = DataLoader(test_set, batch_size=args.batch_size, shuffle=True)

# define fully connected network
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(28*28, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 32)
        self.fc4 = nn.Linear(32, 10)

    def forward(self, x):
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)
        x = F.relu(x)
        x = self.fc4(x)
        output = F.log_softmax(x, dim=1)
        return output
```

```python
def lbfgs_attack_single(model, x, y, l2_reg=1e-5, max_iter=200, lr=1.0, epsilon=0.1):#single to keep runtime down
    epsilon = float(min(epsilon, 0.11))
    
    model.eval()
    x_orig = x.detach().clone()
    x_adv_param = nn.Parameter(x_orig.clone())
    
    optimizer = optim.LBFGS([x_adv_param], lr=lr, max_iter=max_iter, 
                            line_search_fn='strong_wolfe')
    
    def closure():
        optimizer.zero_grad()
        outputs = model(x_adv_param)
        loss_ce = -F.nll_loss(outputs, y)  # maximize CE
        l2_term = ((x_adv_param - x_orig).pow(2).sum(dim=1)).mean()
        loss = loss_ce + l2_reg * l2_term
        loss.backward()
        return loss
    
    try:
        optimizer.step(closure)
    except:
        pass
    
    x_adv = x_adv_param.detach()
    perturbation = torch.clamp(x_adv - x_orig, -epsilon, epsilon)
    x_adv = torch.clamp(x_orig + perturbation, 0.0, 1.0)
    
    return x_adv.detach()


def fgsm_attack(model, x, y, epsilon):
    epsilon = float(min(epsilon, 0.11))
    
    x_adv = x.detach().clone()
    x_adv.requires_grad = True
    
    outputs = model(x_adv)
    loss = F.nll_loss(outputs, y)
    
    model.zero_grad()
    loss.backward()
    
    with torch.no_grad():
        x_adv = x_adv + epsilon * x_adv.grad.sign()
        x_adv = torch.clamp(x_adv, 0.0, 1.0)
    
    return x_adv.detach()


def pgd_attack(model, x, y, epsilon, alpha, iters):
    epsilon = float(min(epsilon, 0.11))
    
    x_orig = x.detach().clone()
    x_adv = x_orig + torch.empty_like(x_orig).uniform_(-epsilon, epsilon)
    x_adv = torch.clamp(x_adv, 0.0, 1.0)
    
    for _ in range(iters):
        x_adv.requires_grad = True
        outputs = model(x_adv)
        loss = F.nll_loss(outputs, y)
        
        model.zero_grad()
        loss.backward()
        
        with torch.no_grad():
            x_adv = x_adv + alpha * x_adv.grad.sign()
            delta = torch.clamp(x_adv - x_orig, -epsilon, epsilon)
            x_adv = torch.clamp(x_orig + delta, 0.0, 1.0).detach()
    
    return x_adv


def cw_attack(model, x, y, c=1.0, kappa=0, max_iter=100, learning_rate=0.01, epsilon=0.1):
    epsilon = float(min(epsilon, 0.11))
    model.eval()
    
    x_orig = x.detach().clone()
    
    # Use tanh space for box constraints
    w = torch.zeros_like(x, requires_grad=True)
    optimizer = optim.Adam([w], lr=learning_rate)
    
    for step in range(max_iter):
        # Map w to valid image space
        x_adv = 0.5 * (torch.tanh(w) + 1.0)
        
        # Project to epsilon ball
        delta = torch.clamp(x_adv - x_orig, -epsilon, epsilon)
        x_adv = torch.clamp(x_orig + delta, 0.0, 1.0)
        
        outputs = model(x_adv)
        
        # C&W loss
        real = outputs.gather(1, y.unsqueeze(1)).squeeze(1)
        other = outputs.clone()
        other.scatter_(1, y.unsqueeze(1), -float('inf'))
        other_max = other.max(1)[0]
        
        loss_adv = torch.clamp(real - other_max + kappa, min=0.0).mean()
        loss_l2 = ((x_adv - x_orig) ** 2).sum(dim=1).mean()
        loss = loss_adv + c * loss_l2
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    with torch.no_grad():
        x_adv = 0.5 * (torch.tanh(w) + 1.0)
        delta = torch.clamp(x_adv - x_orig, -epsilon, epsilon)
        x_adv = torch.clamp(x_orig + delta, 0.0, 1.0)
    
    return x_adv.detach()


def ensemble_attack(model, x, y, epsilon=0.1):

    epsilon = float(min(epsilon, 0.11))
    model.eval()
    
    best_adv = None
    best_loss = -float('inf')
    
    # Only try the 3 STRONGEST attacks. this is not the full attack but a short version that keeps runtime down but still produces a strong output
    attacks_to_try = [
        # PGD-100
        ('pgd_100', lambda: pgd_attack(model, x, y, epsilon, 0.0025, 100)),
        
        # L-BFGS
        ('lbfgs', lambda: lbfgs_attack_single(model, x, y, l2_reg=5e-6, max_iter=150, lr=1.0, epsilon=epsilon)),
        
        # C&W attack
        ('cw', lambda: cw_attack(model, x, y, c=0.3, max_iter=80, learning_rate=0.01, epsilon=epsilon))
    ]# stored attacks in a tuple to try
    
    for attack_name, attack_fn in attacks_to_try:
        try:
            adv_candidate = attack_fn()
            
            with torch.no_grad():
                outputs = model(adv_candidate)
                loss = F.nll_loss(outputs, y, reduction='none').mean().item()
                
                if loss > best_loss:
                    best_loss = loss
                    best_adv = adv_candidate
        except:
            continue
    
    return best_adv if best_adv is not None else x


def adv_attack(model, X, y, device, epsilon=0.1, attack_type='ensemble'):# IF YOU ARE EVALUATING THE ATTACK CHANGE attack_type to 'ensemble' if it is not already that<-------------------------------
    
    if X.dim() == 4:
        X = X.view(X.size(0), -1)
    
    epsilon = float(min(epsilon, 0.11))
    
    if attack_type == 'fgsm':
        # Made so I can test individual attacks
        return fgsm_attack(model, X, y, epsilon)
    
    elif attack_type == 'pgd': #Notes: Basically the iterative version of fgsm
        # Made so I can test individual attacks
        return pgd_attack(model, X, y, epsilon, alpha=0.01, iters=50)
    
    elif attack_type == 'lbfgs':# NOTES: slow but effective. should use for training but only a small amount to keep runtime down
        # Made so I can test individual attacks
        return lbfgs_attack_single(model, X, y, l2_reg=1e-5, max_iter=200, 
                                  lr=1.0, epsilon=epsilon)
    
    elif attack_type == 'cw':
        # Made so I can test individual attacks
        return cw_attack(model, X, y, c=0.5, max_iter=100, 
                        learning_rate=0.01, epsilon=epsilon)
    
    elif attack_type == 'ensemble':
        return ensemble_attack(model, X, y, epsilon=epsilon)
    
    elif attack_type == 'mixed':# Notes: for some reason including pgd and fgsm gives a stronger overall score even though they are very similar
        # FAST mixed strategy for TRAINING <-------------------------------------------------
        # Different attack types helps robustness vs just one type
        rand = torch.rand(1).item()
        if rand < 0.6:
            return fgsm_attack(model, X, y, epsilon)
        elif rand < 0.9:
            return pgd_attack(model, X, y, epsilon, alpha=epsilon/7, iters=7)
        else:
            return lbfgs_attack_single(model, X, y, l2_reg=1e-5, max_iter=20, 
                                      lr=1.0, epsilon=epsilon)
    
    # DEFAULT: Use mixed for safety
    rand = torch.rand(1).item()
    if rand < 0.6:
        return fgsm_attack(model, X, y, epsilon)
    elif rand < 0.9:
        return pgd_attack(model, X, y, epsilon, alpha=epsilon/7, iters=20)
    else:
        return lbfgs_attack_single(model, X, y, l2_reg=1e-5, max_iter=20, 
                                  lr=1.0, epsilon=epsilon)


# Define loss function
criterion = nn.NLLLoss()


def train(args, model, device, train_loader, optimizer, epoch):
    """Improved training with progressive epsilon and stronger attacks."""
    model.train()
    
    # Progressive epsilon schedule
    epsilon_schedule = min(0.1, 0.05 + (epoch - 1) * 0.01)
    epsilon_schedule = min(epsilon_schedule, 0.1)
    
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        data = data.view(data.size(0), -1)
        
        # Generate STRONGER adversarial examples during training
        rand = torch.rand(1).item()
        if rand < 0.3:  # 30% FGSM (fast)
            adv_data = fgsm_attack(model, data, target, epsilon_schedule)
        elif rand < 0.9:  # 60% PGD-10 (stronger than before)
            adv_data = pgd_attack(model, data, target, epsilon_schedule, 
                                 alpha=epsilon_schedule/10, iters=10)
        else:  # 10% L-BFGS (very strong)
            adv_data = lbfgs_attack_single(model, data, target, 
                                          l2_reg=1e-5, max_iter=30, 
                                          lr=1.0, epsilon=epsilon_schedule)
        
        optimizer.zero_grad()
        
        # Better balance: 25% clean, 75% adversarial
        out_clean = model(data)
        out_adv = model(adv_data)
        
        loss = 0.25 * criterion(out_clean, target) + 0.75 * criterion(out_adv, target)
        loss.backward()
        optimizer.step()


#predict function
def eval_test(model, device, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            data = data.view(data.size(0),28*28)
            output = model(data)
            test_loss += F.nll_loss(output, target, size_average=False).item()
            pred = output.max(1, keepdim=True)[1]
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)
    test_accuracy = correct / len(test_loader.dataset)
    return test_loss, test_accuracy

def eval_adv_test(model, device, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    for data, target in test_loader:
        data, target = data.to(device), target.to(device)
        data = data.view(data.size(0),28*28)
        adv_data = adv_attack(model, data, target, device=device)
        with torch.no_grad():
            output = model(adv_data)
            test_loss += F.nll_loss(output, target, size_average=False).item()
            pred = output.max(1, keepdim=True)[1]
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)
    test_accuracy = correct / len(test_loader.dataset)
    return test_loss, test_accuracy


def train_model():
    print("Initializing model...")
    model = Net().to(device)
    optimizer = optim.Adam(model.parameters(), lr=args.lr)
    
    print(f"Starting training for {args.epochs} epochs...")
    total_start = time.time()
    
    for epoch in range(1, args.epochs + 1):
        epoch_start = time.time()
        
        train(args, model, device, train_loader, optimizer, epoch)
        
        trn_loss, trn_acc = eval_test(model, device, train_loader)
        adv_loss, adv_acc = eval_adv_test(model, device, test_loader)
        
        epoch_time = int(time.time() - epoch_start)
        
        print(f"Epoch {epoch}: {epoch_time}s, "
              f"trn_loss: {trn_loss:.4f}, trn_acc: {100*trn_acc:.2f}%, "
              f"adv_loss: {adv_loss:.4f}, adv_acc: {100*adv_acc:.2f}%")
    
    total_time = int(time.time() - total_start)
    print(f"\nTotal training time: {total_time}s ({total_time/60:.1f} minutes)")
    
    print('\n' + '='*80)
    print('Training completed! Model saved as: {}.pt'.format(id_))
    print('='*80)
    print('\nTo evaluate your attack and defence scores:')
    print('  1. Test your attack:  python Evaluate_attack.py')
    print('  2. Test your defence: python Evaluate_defence.py')
    print('\nMake sure to:')
    print('  - Update attack method in Evaluate_attack.py')
    print('  - Update model file path in Evaluate_defence.py')
    print('  - Verify epsilon constraint < 0.11 (see p_distance output below)')
    print('='*80 + '\n')
    
    torch.save(model.state_dict(), f"{id_}.pt")
    return model


def p_distance(model, train_loader, device):
    """Measure maximum perturbation using mixed attack."""
    p = []
    
    print("Computing epsilon constraint...")
    for data, target in train_loader:
        data, target = data.to(device), target.to(device)
        data = data.view(data.size(0), -1)
        
        # Use the same mixed attack as training
        adv_data = adv_attack(model, data, target, device,
                             epsilon=0.1, attack_type='mixed')
        
        diff = torch.abs(data - adv_data)
        p_vals = diff.view(diff.size(0), -1).max(dim=1)[0]
        p.extend(p_vals.cpu().numpy().tolist())
    
    max_p = max(p)
    print(f"epsilon p: {max_p:.6f}")
    
    if max_p > 0.11:
        print(f"WARNING: epsilon constraint violated! {max_p:.6f} > 0.11")
    else:
        print(f"✓ Epsilon constraint satisfied: {max_p:.6f} <= 0.11")
    
    return max_p
```
